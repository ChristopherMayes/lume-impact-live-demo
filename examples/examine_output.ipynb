# Examine results from the LUME-Impact Live Demo

Every evaluation writes a summary JSON file which contains top level:
- `isotime`: the time that the EPICS snapshot data was taken
- `config`: basic configuration with base models
- `inputs`: input settings for these models
- `outputs`: simple outputs at the end of the run, plus the archive file
- `pv_mapping_dataframe`: Table of EPICS <-> Simulation mapping data
